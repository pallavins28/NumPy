2. How does broadcasting work in NumPy?
--> Broadcasting is a powerful feature in NumPy that allows operations between arrays of different shapes without explicitly duplicating data. This makes computations more efficient and memory-friendly.

When performing element-wise operations between arrays, NumPy applies the following rules to make them compatible:

1.If the arrays have different dimensions, prepend 1s to the smaller arrayâ€™s shape until they match.

2.An axis with size 1s can be stretched to match the corresponding dimension of the larger array.

3.If, after applying rules 1 and 2, the shapes match, element-wise operations are performed. Otherwise, NumPy raises an error.

**Examples of Broadcasting**

**1. Scalar and Array Operations**

A scalar is stretched across all elements of the array.

    import numpy as np
    arr = np.array([1, 2, 3])
    result = arr + 5  # Broadcasting 5 to match arr's shape
    print(result)  # Output: [6 7 8]

**Broadcasting Between 1D and 2D Arrays**
A 1D array can be broadcasted to a 2D array.

    arr1 = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2,3)
    arr2 = np.array([10, 20, 30])  # Shape (3,) -> Broadcasted to (1,3)

    result = arr1 + arr2
    print(result)

    # Output:
    # [[11 22 33]
    #  [14 25 36]]
Here, arr2 is reshaped to (1, 3), then broadcasted to (2, 3), matching arr1.

**Broadcasting in Higher Dimensions**

If dimensions donâ€™t match, NumPy expands the lower-dimensional array.

    arr1 = np.array([[1], [2], [3]])  # Shape (3,1)
    arr2 = np.array([10, 20, 30])  # Shape (3,) -> Broadcasted to (1,3)

    result = arr1 + arr2
    print(result)

    # Output:
    # [[11 21 31]
    #  [12 22 32]
    #  [13 23 33]]
Here, arr1 reshapes to (3,1), and arr2 reshapes to (1,3), then both become (3,3).

**When Broadcasting Fails**

If the shapes are incompatible and cannot be aligned, NumPy raises an error.

    arr1 = np.array([[1, 2], [3, 4]])  # Shape (2,2)
    arr2 = np.array([1, 2, 3])  # Shape (3,) - Incompatible

    result = arr1 + arr2  # This will raise a ValueError
This fails because (2,2) and (3,) cannot be broadcasted together.

6. What are the differences between NumPy arrays and Python lists?

--> Differences Between NumPy Arrays and Python Lists

Performance: NumPy arrays are faster because they use optimized C implementations and store data in contiguous memory, whereas Python lists store elements as individual objects, leading to overhead.

Memory Efficiency: NumPy arrays use fixed-type storage, consuming less memory, while Python lists store heterogeneous data types, requiring more memory per element.

Element Type: NumPy arrays are homogeneous (all elements must be of the same type), whereas Python lists can store heterogeneous data types.

Operations: NumPy supports vectorized operations (e.g., arr * 2), whereas Python lists require loop-based operations (e.g., [x * 2 for x in list]).

Multi-dimensional Support: NumPy arrays can handle multi-dimensional structures (matrices, tensors), while Python lists only support 1D nested lists.

Built-in Functions: NumPy provides optimized mathematical functions (such as sum, mean, standard deviation), whereas Python lists lack direct numerical operations.

Indexing & Slicing: NumPy arrays allow efficient and flexible slicing (arr[1:3]), whereas Python lists support slicing but are less efficient.

Broadcasting: NumPy arrays support broadcasting, enabling element-wise operations across different shapes, while Python lists require explicit loops.

Library Dependencies: NumPy requires installation (numpy library), whereas Python lists are a built-in feature of Python.

Use Case: NumPy arrays are ideal for large-scale numerical computations, while Python lists are more suitable for general-purpose storage and mixed data types. ðŸš€

Examples of Key Differences

1.Performance Comparison

    import numpy as np
    import time

    size = 1000000  # Large array size

    # Using a Python list
    py_list = list(range(size))
    start = time.time()
    py_result = [x * 2 for x in py_list]
    print("Python List Time:", time.time() - start)

    # Using a NumPy array
    np_array = np.arange(size)
    start = time.time()
    np_result = np_array * 2  # Vectorized operation
    print("NumPy Array Time:", time.time() - start)
ðŸ”¹ NumPy is significantly faster due to its optimized C implementation.

2.Memory Efficiency

    import sys
    py_list = [1, 2, 3, 4, 5]
    np_array = np.array(py_list)

    print("Python List Size:", sys.getsizeof(py_list))
    print("NumPy Array Size:", np_array.nbytes)
ðŸ”¹ NumPy uses less memory because it stores data in a contiguous memory block.

3.Multi-dimensional Support

    np_matrix = np.array([[1, 2, 3], [4, 5, 6]])
    print(np_matrix.shape)  # (2, 3)
ðŸ”¹ Python lists would require nested lists for this structure.

8. What does the term "vectorized operation" mean in NumPy?
-->A vectorized operation in NumPy refers to performing operations on entire arrays (vectors, matrices) without using explicit loops. It leverages low-level optimizations in C to execute computations faster and more efficiently than traditional Python loops.

Key Advantages of Vectorized Operations ðŸš€
âœ… Faster Execution: Eliminates the overhead of Python loops by leveraging SIMD (Single Instruction, Multiple Data) operations.

âœ… Memory Efficient: Works on contiguous memory blocks, reducing overhead.

âœ… Concise Code: Simplifies complex operations with minimal syntax.

Example: Vectorized vs. Non-Vectorized Operations
ðŸ”´ Without Vectorization (Using Python Loop)

    import numpy as np

    # Creating two lists
    list1 = [1, 2, 3, 4, 5]
    list2 = [10, 20, 30, 40, 50]

    # Element-wise multiplication using loop
    result = [a * b for a, b in zip(list1, list2)]
    print(result)  # Output: [10, 40, 90, 160, 250]

With Vectorization (Using NumPy)

    arr1 = np.array([1, 2, 3, 4, 5])
    arr2 = np.array([10, 20, 30, 40, 50])

    # Vectorized multiplication
    result = arr1 * arr2
    print(result)  # Output: [10 40 90 160 250]

21. Why is NumPy's array slicing faster than Python's list slicing?
-->  Why is NumPy's Array Slicing Faster Than Python's List Slicing?

NumPy arrays are significantly faster than Python lists, especially when slicing and performing operations. This performance difference is due to several key optimizations in NumPy.

NumPy Uses Views Instead of Copies

NumPy slicing returns a "view" (same memory block) rather than a copy, avoiding extra memory usage.

Python list slicing creates a new list (deep copy), which is slower and consumes more memory.

Example: NumPy View vs. Python List Copy

    import numpy as np

    # NumPy Array
    arr = np.array([1, 2, 3, 4, 5])
    slice_arr = arr[1:4]  # View (no extra memory used)
    slice_arr[0] = 99  # Modifying slice modifies original array

    print(arr)  # Output: [ 1 99  3  4  5]

    # Python List
    lst = [1, 2, 3, 4, 5]
    slice_lst = lst[1:4]  # Creates a new list (copy)
    slice_lst[0] = 99  # Modifying slice does NOT affect the original list

    print(lst)  # Output: [1, 2, 3, 4, 5]
Result: NumPy is more memory-efficient, whereas Python lists require extra copying.
